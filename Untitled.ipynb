### Функция варианта:

$\large f(x, y) = 5 (x+y)^2 + (x-2)^2 \rightarrow min$

```python
def f(x, y):
    return 5*(x + y)**2 + (x - 2)**2
```

### Ограничения функции:

a) $x + y \geq 1$

б) $ x = y$

```python
# x + y >= 1
def constraint_alpha(x, y):
    return - x - y + 1

# x = y
def constraint_beta(x, y):
    return x - y

constraints = [constraint_alpha, constraint_beta]
```

## Имплементация алгоритма

$max(0, c(x, y)^2$

```python
def penalty_function_powMax(x, y, constraint_function):
    return np.power(np.max([0, constraint_function(x, y)]), 2)
```

```python
def penalty_square(x, y, constraint_function):
    return constraint_function(x, y)**2
```

```python
def function_with_penalty(X, constraints, penalty_rate, pf):
    return f(*X) + penalty_rate * np.sum(np.array(
        [pf(*X, constraint) for constraint in constraints]
    ))
```

```python
def penalty_function_method(function, constraints, initial_approximixation, precision, penalty_rate, penalty_descent_rate, penalty_function):
    
    history = [initial_approximixation.copy()]
    current = initial_approximixation.copy()
    
    while True:
        fwp_partial = lambda x: function_with_penalty(x, constraints, penalty_rate, penalty_function)
        current = minimize(fwp_partial, current.copy(), method='Powell').x
        history.append(current.copy())
        
        if fwp_partial(current) - function(*current) <= precision:
            return current, history
        
        penalty_rate += penalty_rate * penalty_descent_rate
```

```python
columns = [
    'Функция штрафа',
    'Нач. штраф',
    'Измен. коэф. штрафа',
    'Нач. точка',
    'Точность',
    'Кол-во итераций',
    'Найденная точка',
    'Знач. в точке'
]
```

```python
table = pd.DataFrame(columns=columns)
```

```python
def calculate_and_fill_table(func, const, x, prec, pnlt, r_pnlt, pnlt_f, pnlt_f_name):
    point, hist = penalty_function_method(func, const, x, prec, pnlt, r_pnlt, pnlt_f)
    iter_count = len(hist)

    row = len(table)
    table.loc[row] = [
        pnlt_f_name,
        pnlt,
        r_pnlt,
        x,
        prec,
        iter_count,
        np.round(point, 4),
        np.round(f(*point), 5)        
    ]
```

```python
for param in params:
    calculate_and_fill_table(*param)
```

## Выводы

Как можно видеть по таблице и графикам:
1) На кол-во итераций сильно влияет начальное значение штрафа/штрафного коэффициента. Это связано с тем, что при попытке оптимизации в "штрафную область", значение целевой функции сильно штрафуются и алгоритм вынужден соблюдать ограничения.

2) Аналогичная ситуация обстоит с коэффициентом изменения штрафа. Чем он выше, чем сильнее алгоритм будет штрафоваться за выходы из "дозволенной" области.

3) Чем дальше начальная точка, тем с большей "увереностью" алгоритм завершит задачу за один шаг. Скорее всего это особенность только данного сочетания ограничений и функции

4) Сниженние точности приводит к ускорению алгоритма.
